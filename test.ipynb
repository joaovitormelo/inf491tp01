```python
from microsynth import Instrumento
```

```python
from IPython.display import Audio
import numpy as np

default_sr = 44100

# Create a basic instrument (sine wave)
instrumento = Instrumento(
    nome="Sintetizador Simples",
    forma="senoide",
    n_harm=8,
    adsr_params=(0.01, 0.1, 0.4, 0.2),  # Attack, Decay, Sustain, Release (in seconds)
    fase=0.0,
    unidade_fase='graus'
)

# Generate a simple note (Middle C = 262 Hz, duration = 1 second)
nota = instrumento.gerar_nota(f=262, dur=1.0, sr=default_sr, amp=0.8)

# Display with native IPython player
Audio(data=nota, rate=default_sr)
```
