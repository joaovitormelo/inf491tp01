```python
from microsynth import Instrumento
```

```python
from IPython.display import Audio
import numpy as np

default_sr = 44100

instrumento = Instrumento(
    nome="Sintetizador Teste",
    forma="senoide",
    n_harm=8,
    adsr_params=(0.01, 0.1, 0.4, 0.2),
    fase=0.0,
    unidade_fase='graus'
)

nota = instrumento.gerar_nota(f=262, dur=1.0, sr=default_sr, amp=0.8)

Audio(data=nota, rate=default_sr)
```

```python
piano = Instrumento(
    nome='Piano-like',
    forma='triangular',      # onda naturalmente rica em harmônicos
    n_harm=10,          # quantidade moderada de harmônicos
    adsr_params=(
        0.01,           # Ataque rápido
        0.2,            # Decaimento moderado
        0.4,            # Sustain baixo
        0.3             # Liberação moderada
    ),
    fase=0,
    unidade_fase='graus',
    am_params={
        "f_mod": 800,
        "I": 0.1
    },
    fm_params= None,
    # fm_params={
    #     "f_m": 220,
    #     "I": 0.5, "tipo_fm": "const"
    # },
    filtro_params={
        'cutoff_high': 3000  # passa-baixa leve
    },
    efeito_params=None
)
```

```python
from microsynth import midi2freq

t, y = piano.gerar_nota(
    f=midi2freq(60),
    dur=1.0,
    retorna_t=True
)

Audio(data=y, rate=default_sr)
```
